```python
df = pd.read_csv('fifa21.csv', low_memory=False)
```

```python
df
```

```python
#Cleaning the dataset
df['Loan Date End'].isnull().sum()
```

```python
#Removing the "\n\n\n\n" value from the Club
df['Club'] = df['Club'].str.replace('\n\n\n\n', '')
```

```python
#Fixing Datatypes and replacing values
df['Hits'] = df['Hits'].fillna(0)
df['Hits'] = df['Hits'].astype(str)
def convert_K_to_numerical(value):
    if 'K' in value:
        return float(value.replace('K', '')) * 1000
    else:
        return float(value)

# Apply the conversion function to the 'Hits' column
df['Hits'] = df['Hits'].apply(convert_K_to_numerical)
df['Hits'] = df['Hits'].astype(float)
```

```python
#Removing PR from nationality column
df['Nationality'] = df['Nationality'].str.replace('PR', '')
```

```python
#Handling NaN by replacing with mean values
nonadf = df[df['Hits'].notna()]
mean = nonadf['Hits'].mean()

df['Hits'] = df['Hits'].replace(0.0, mean)
df
```

```python
#Renaming playerUrl column
df = df.rename(columns={'playerUrl': 'Player URL'})
```

```python
df = df.rename(columns={'photoUrl': 'Photo URL'})
```
